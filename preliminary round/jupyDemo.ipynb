```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Importing the dataset
#dataset = open('C:/Users/DCS/Desktop/Tbrain/Code/train.csv')
dataset = pd.read_csv('train.csv',encoding='big5')
```

```python
# Taking care of missing data[](以0替代)
dataset["ANNUAL_PREMIUM_AMT"].fillna("0", inplace = True) 
dataset["INSD_LAST_YEARDIF_CNT"].fillna("0", inplace = True) 
dataset["L1YR_C_CNT"].fillna("0", inplace = True) 

X = dataset.iloc[:, 48:73].values
y = dataset.iloc[:, 131].values

# Taking care of missing data[](Annual_Income_AMT, 平均值)
from sklearn.preprocessing import Imputer
imputer=Imputer(missing_values="NaN", strategy="mean" )
dataset["ANNUAL_INCOME_AMT"]=imputer.fit_transform(dataset[["ANNUAL_INCOME_AMT"]]).ravel()

# Taking care of missing data[](BMI, 平均值)
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean')
dataset["BMI"]=imputer.fit_transform(dataset[["BMI"]]).ravel()

# Taking care of missing data[](Termination_Rate,以0替代)
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'most_frequent')
dataset["TERMINATION_RATE"]=imputer.fit_transform(dataset[["TERMINATION_RATE"]]).ravel()
```
